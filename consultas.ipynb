```python
SELECT
    TO_CHAR(pickup_ts, 'YYYY-MM') AS year_month,
    COUNT(*) AS total_trips
FROM gold.fct_trips
WHERE pickup_ts >= '2024-01-01' AND pickup_ts < '2025-01-01'
GROUP BY year_month
```

```python
SELECT
    TO_CHAR(pickup_ts, 'YYYY-MM') AS year_month,
    service_type,
    COUNT(*) AS total_trips
FROM gold.fct_trips
WHERE pickup_ts >= '2024-01-01' AND pickup_ts < '2025-01-01'
GROUP BY year_month, service_type
ORDER BY year_month, service_type;
```

```python
SELECT
    z.zone_name,
    z.borough,
    COUNT(*) AS total_trips
FROM gold.fct_trips t
JOIN gold.dim_zone z ON t.pu_zone_key = z.zone_key
WHERE t.pickup_ts >= '2024-01-01' AND t.pickup_ts < '2025-01-01'
GROUP BY z.zone_name, z.borough
ORDER BY total_trips DESC
LIMIT 10;
```
